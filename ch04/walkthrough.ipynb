```python
def compose(f1, f2):
    def new_function(input):
        return f1(f2(input))
    return new_function
```

```python
def polygon_map(transformation, polygons):
    return [[transformation(vertex) for vertex in triangle] for triangle in polygons]
```

```python
from vectors import scale
def scale_by(scalar):
    def new_function(v):
        return scale(scalar, v)
    return new_function
```
